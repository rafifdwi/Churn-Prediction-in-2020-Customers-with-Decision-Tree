## Encoding Data dengan LabelEncoder

```python
data_encod = data_train.drop(columns=['state','area_code'],axis=1)
data_encod.head()
```

```python
from sklearn.preprocessing import LabelEncoder
#Convert all the non-numeric columns to numerical data types
for column in data_encod.columns:
	if data_encod[column].dtype == np.number: continue
	# Perform encoding for each non-numeric column
	data_encod[column] = LabelEncoder().fit_transform(data_encod[column])
print(data_encod.describe())
```

```python
data_train.head()
```

```python
data_encod.head()
```

### encoding pada data_test dengan mapping

```python
dictInt = { "no" : 0, "yes" : 1 }
dictVoice = { "no" : 0, "yes" : 1  }
data_test['international_plan'] = data_test['international_plan'] .map(dictInt)
data_test['voice_mail_plan'] = data_test['voice_mail_plan'] .map(dictVoice)
```

```python

x_train = data_encod.drop('churn',axis=1)
y_train = data_encod['churn']
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_log = LogisticRegression().fit(x_train,y_train)
print(model_log)
```

```python
from sklearn.metrics import classification_report
y_train_pred = model_log.predict(x_train)
# Print classification report
print('Classification Report Training Model (Logistic Regression) :')
print(classification_report(y_train, y_train_pred))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
error = []

for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)

    pred_i = knn.predict(x_train)
    error.append(np.mean(pred_i != y_train))

plt.figure()
plt.plot(range(1, 40), error, color='red', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Rata-Rata Error terhadap nilai K')
plt.xlabel('Nilai K')
plt.ylabel('Rata-Rata Error')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=30)
knn.fit(x_train, y_train)
score_knn = knn.score(x_train,y_train)
print(knn.score(x_train,y_train))
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
score_knn = knn.score(x_train,y_train)
print(knn.score(x_train,y_train))
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
score_knn = knn.score(x_train,y_train)
print(knn.score(x_train,y_train))
```

```python
model = KNeighborsClassifier()
model_knn = model.fit(x_train, y_train)
y_train_pred = model_knn.predict(x_train)
# Print classification report
print('Classification Report Training Model (KNN) :')
print(classification_report(y_train, y_train_pred))
```
